```python
import seaborn as sns
import pandas as pd

#  seaborn 
sns.set()

tips = pd.read_csv('./data/tips.csv') 

# x - 
# y - 
# col - 
# hue -  

sns.relplot(x="total_bill", y="tip", col="time",
            hue="smoker", style="smoker", size="size",
            data=tips);
```
