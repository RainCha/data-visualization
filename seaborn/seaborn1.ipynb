```python
import seaborn as sns
import pandas as pd

# 设置并使用 seaborn 默认的主题、尺寸大小以及调色板
sns.set()

tips = pd.read_csv('./data/tips.csv') 

# x - 横坐标
# y - 纵坐标
# col - 以什么字段来分子图列
# hue - 以什么字段 区分颜色

sns.relplot(x="total_bill", y="tip", col="time",
            hue="smoker", style="smoker", size="size",
            data=tips);
```
